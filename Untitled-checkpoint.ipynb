```python
import pandas as pd

RezL=pd.read_csv('Harv_I_RezL.CSV',delimiter=';')

column_names = ["DateTime"]

M1 = pd.DataFrame(columns = column_names)

M1['DateTime'] = pd.date_range(start=pd.Timestamp('9/11/2020 00:00:00'), end=pd.Timestamp('11/23/2020 00:00:00'), freq = '1S')

RezL['DateTime']= pd.to_datetime(RezL['DateTime'], format ="%d.%m.%Y %H:%M:%S")
M1['DateTime']= pd.to_datetime(M1['DateTime'], format ="%d.%m.%Y %H:%M:%S")
```

```python
# RezL_new = RezL_new.drop(RezL_new[(RezL_new['DateTime'].dt.month == 11) & (RezL_new['DateTime'].dt.day > 22)].index)

RezL_new = pd.merge(M1,RezL, how='outer', on = 'DateTime')

RezL_new = RezL_new.fillna(method='ffill')

# RezL_new.drop(RezL_new.tail(2).index,inplace=True)

RezL_new = RezL_new.drop_duplicates(subset=['DateTime']) 

RezL_new = RezL_new.reset_index()

# print(RezL_new)

RezL_new = RezL_new.iloc[::5, :]

RezL_new = RezL_new.reset_index()

print(RezL_new)

# # with pd.option_context('display.max_rows', None, 'display.max_columns', None):
# #     print(RezL_new[0:1000])
```
